```python
# Test Train Data
X = df.iloc[:, 0:7]
y = df['category']
```

```python
# Spliting the dataset intro train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 32)
```

```python
# Checking the shape of the split
print("X train: {}".format(X_train.shape))
print("X test: {}".format(X_test.shape))
print("y train: {}".format(y_train.shape))
print("y test: {}".format(y_test.shape))
```

```python
# K-means Clustering

# Finding the optimum value of n_clusters
from sklearn.cluster import KMeans


# Fitting the model using the above optimum value
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
               max_iter = 300, random_state = 32)
kmeans.fit(X_train)

y_pred = kmeans.predict(X_test)
```

```python
# Make Predictions using it
new = np.array([[13.00,  11, 86, 4.9, 3.2, 2.12, 4.22]])

pred = kmeans.predict(new)
print(pred)
```
